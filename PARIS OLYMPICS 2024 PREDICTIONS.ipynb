```python
data=athletes.dropna(subset=["Medal"])
data.drop_duplicates(subset=["Team", "NOC", "Games", "Event", "Medal", "City", "Sport","Year"], inplace=True)
```

```python
new_df=data[data["region"]=="Pakistan"]
pakistan=new_df.groupby("Year").count()["Medal"].reset_index()
pakistan
```

```python
plt.plot(pakistan['Year'], pakistan['Medal'], label='Medal Count', color="green")
plt.xlabel('Year')
plt.ylabel('Medal Count')
plt.title('Medal Count Over Years for Pakistan', fontweight="bold", color="g")
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
new_df=data[data["region"]=="USA"]
usa=new_df.groupby("Year").count()["Medal"].reset_index()

plt.plot(usa['Year'], usa['Medal'], label='Medal Count', color="red")
plt.xlabel('Year')
plt.ylabel('Medal Count')
plt.title('Medal Count Over Years for USA',fontweight="bold", color="red")
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
